```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pycaret.regression as pycReg
import pycaret.classification as pycClf
```

```python
df = pd.read_csv('data/preprocessed.csv')
df.shape
```

```python
#setup model
model = pycClf.setup(data = df, 
                    target = 'target',
                    train_size=0.8,
                    normalize=False,
                    verbose=False,
                    numeric_features=list(df.drop(['target'], axis=1).columns),
                    silent = True)
```

```python
# #compare models
# pycReg.compare_models(sort = "RMSE")
```

```python
#create model
rf = pycReg.create_model('rf', verbose=True)
```

```python
# #tune model
# catB = pycReg.tune_model(catB, optimize = "RMSE", n_iter = 500)
```

```python
import shap
explainer = shap.TreeExplainer(rf)
my_shap_values = explainer.shap_values(df.drop(['target'], axis=1).values)
```

```python
shap.summary_plot(my_shap_values, df.drop(['target'], axis=1).values, show=None)
plt.savefig('figures/tree_shap_impact_overall.png')
```

```python
#taking the shape value only for the target class.
my_tar = list(df['target'])
true_shap = np.zeros((df.drop(['target'], axis=1).shape[0], df.drop(['target'], axis=1).shape[1]))

for i in range(len(my_tar)):
    ind = my_tar[i]
    temp_shap = my_shap_values[ind-1]
    true_shap[i,:] = temp_shap[i,:]
```

```python
shap.summary_plot(true_shap, df.drop(['target'], axis=1).values, show=None)
plt.savefig('figures/tree_shap_impact_prediction.png')
```

```python
shap.summary_plot(true_shap, df.drop(['target'], axis=1).values, plot_type='bar', show=None)
plt.savefig('figures/tree_shap_contribution.png')
```

```python
#Adding the shap value of one-hot encoded column
new_shap = np.zeros((117,27))

new_shap[:,:22] = true_shap[:,:22]
new_shap[:,22] = np.sum(true_shap[:,22:24], axis=1)
new_shap[:,23] = np.sum(true_shap[:,24:27], axis=1)
new_shap[:,24] = np.sum(true_shap[:,27:31], axis=1)
new_shap[:,25] = np.sum(true_shap[:,31:36], axis=1)
new_shap[:,26] = np.sum(true_shap[:,36:38], axis=1)
```

```python
shap.summary_plot(new_shap, plot_type='bar', show=None)
plt.savefig('figures/tree_shap_contribution_exact.png')
```
